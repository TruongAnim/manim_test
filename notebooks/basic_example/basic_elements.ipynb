# Rendering with Jupyter

As we saw in the tutorial, we can render a scene using the magic method:

> `%%manim -qm <SCENE-NAME>`

However, it is good practice to use the flags `--disable_caching` and `-v WARNING`, we will explain why in Chapter 3.

```python
%%manim -qm --disable_caching -v WARNING Example

class Example(Scene):
    def construct(self):
        self.play(Write(Square()))
        self.wait()
```

However this can be very tedious, so we can shorten this as follows

```python
render_as_image = "-s --disable_caching -v WARNING Example"
render_as_video = "-qm --disable_caching -v WARNING Example"
```

The only limitation of this technique is that all your scenes must be named "Example".

```python
%%manim $render_as_image

config.media_embed = True # <- Use this if you render image

class Example(Scene):
    def construct(self):
        self.play(Write(Square()))
        self.wait()
```

```python
%%manim $render_as_video

class Example(Scene):
    def construct(self):
        self.play(Write(Square()))
        self.wait()
```

--------------
SETUP METHOD
--------------

In future scenes we will need the scenes to have certain default settings, in general, the screen coordinates, for this we can use the setup method.

This method is the same as `construct`, but it is executed <u style="color:orange"><b>before</b></u>, so we can use it in the same way.

```python
class MyScene(Scene):
    def setup(self):
        # We'll explain the NumberPlane class in Chapter 14
        self.number_plane = NumberPlane(axis_config={"include_numbers": True})
        self.add(self.number_plane)
```

```python
%%manim $render_as_image

config.media_embed = True
config.background_color = GREY_E

class Example(MyScene): # <--- MyScene instead Scene
    def construct(self):
        pass
```
