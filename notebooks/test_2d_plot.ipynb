```python
from manim import *
config.set_width = "80%"
```

```python
%%manim -qm Test2DPlot

class Test2DPlot(Scene):
    def setup(self):
        axes = Axes(
            x_range=(0, 10, 1),
            y_range=(0, 100, 10),
            x_length=5,
            y_length=5,
            axis_config={
                "include_numbers":True,
                
            }
        )
        graph = axes.plot(lambda x: x**2, x_range=[0,10])
        self.add(axes, graph)
        x = [i for i in range(10)]
        dots = [Dot(axes.c2p(i, i**2), color=RED) for i in x]
        self.add(*dots)
        
```
